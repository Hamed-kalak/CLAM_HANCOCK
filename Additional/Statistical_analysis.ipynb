```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests
from scipy.stats import shapiro, levene, ttest_ind
```

# This part does the t-test and checks for the assumptions as well

```python
sns.set(style="white")
fig, ax = plt.subplots(1, 2, figsize=(18, 7))

test_values_all_tasks = []
labels = []

for model_name, files in loc_dict.items():
    data = pd.read_csv(files['test_auc'])
    test_values = data['test_auc'].values
    test_values_all_tasks.append(test_values)
    labels.append(model_name)

data_combined = pd.DataFrame({label: values for label, values in zip(labels, test_values_all_tasks)})
melted_data = data_combined.melt(var_name='Model', value_name='test_auc')

color_palette = ["#aec7e8", "#aec7e8", "#aec7e8", "#aec7e8","#ff9896","#ff9896"]

# boxplot
sns.boxplot(x='Model', y='test_auc', data=melted_data, ax=ax[0], palette=color_palette, flierprops=dict(marker='o', markersize=5))

# Annotate means on the boxes
means = data_combined.mean()
for i, mean in enumerate(means):
    ax[0].text(i, mean, f'{mean:.2f}', 
               horizontalalignment='center', color='white', fontsize=14, verticalalignment='bottom', bbox=dict(facecolor='black', alpha=0.3))

ax[0].set_ylabel('Test AUC', fontsize=16)
ax[0].set_xlabel('', fontsize=16)
ax[0].tick_params(axis='x', rotation=45, labelsize=14)
ax[0].tick_params(axis='y', labelsize=14)
ax[0].yaxis.grid(True)
ax[0].xaxis.grid(False) 

# Check normality assumption
print("Normality Test --> (p-values):")
normality_results = {label: shapiro(values)[1] for label, values in zip(labels, test_values_all_tasks)}
for label, p_val in normality_results.items():
    if p_val < 0.01:
        print(f"{label}: p = {p_val:.4f} (Reject normality assumption --> meh)")
    else:
        print(f"{label}: p = {p_val:.4f} (Fail to reject normality assumption --> good)")

# Check homogeneity of variances
levene_stat, levene_p_val = levene(*test_values_all_tasks)
print("omogeneity of Variances:")
if levene_p_val < 0.05:
    print(f"p = {levene_p_val:.4f} (Reject homogeneity of variances assumption  --> meh)")
else:
    print(f"p = {levene_p_val:.4f} (Fail to reject homogeneity of variances assumption  --> good)")

# Calculate pairwise p-values
p_values = np.zeros((len(labels), len(labels)))
for i in range(len(labels)):
    for j in range(len(labels)):
        _, p_val = ttest_ind(test_values_all_tasks[i], test_values_all_tasks[j], equal_var=False)
        p_values[i, j] = p_val

# Correcting for multiple test
p_values_flat = p_values.flatten()
corrected_p_values = multipletests(p_values_flat, alpha=0.05, method='holm')[1]
p_values_corrected = corrected_p_values.reshape(len(labels), len(labels))
np.fill_diagonal(p_values_corrected, 1.0)
print("\nHolm-Bonferroni:")
for i in range(len(labels)):
    for j in range(i + 1, len(labels)):
        print(f"{labels[i]} vs {labels[j]}: original p = {p_values[i, j]:.4e}, corrected p = {p_values_corrected[i, j]:.4e}")

significance_level_1 = 0.05
significance_level_2 = 0.01
significance_level_3 = 0.001
significance_mask = np.full_like(p_values_corrected, '', dtype=object)
for i in range(len(labels)):
    for j in range(len(labels)):
        if p_values_corrected[i, j] < significance_level_3:
            significance_mask[i, j] = '**'
        elif p_values_corrected[i, j] < significance_level_2:
            significance_mask[i, j] = '*'
        elif p_values_corrected[i, j] < significance_level_1:
            significance_mask[i, j] = ''
        else:
            significance_mask[i, j] = ''

annotated_p_values = np.empty_like(p_values_corrected, dtype=object)
for i in range(len(labels)):
    for j in range(len(labels)):
        annotated_p_values[i, j] = f"{p_values_corrected[i, j]:.2e}{significance_mask[i, j]}"

# Create a heatmap f
mask = np.triu(np.ones_like(p_values_corrected, dtype=bool), k=1)  # Mask the upper triangle, keep diagonal
sns.heatmap(p_values_corrected, mask=mask, annot=annotated_p_values, fmt="", cmap="Blues_r",
            xticklabels=labels, yticklabels=labels, ax=ax[1],
            linecolor='white', linewidths=0)
cbar = ax[1].collections[0].colorbar
cbar.ax.set_ylabel('p-value', fontsize=14) 
ax[1].tick_params(axis='x', rotation=45, labelsize=14)
ax[1].tick_params(axis='y', labelsize=14)
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45, ha="right")

plt.tight_layout()
plt.savefig(f'/data/ve59kedo/thesis/results/th3_th4_localization.png', dpi=300)
plt.show()
```
